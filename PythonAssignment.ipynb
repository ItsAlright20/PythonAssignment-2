**‚ùìQ11. Write a function calculate_HRR(max_bpm, resting_bpm, intensity=0.7) that returns Heart Rate
Reserve zone.**

```python
def calculate_HRR(max_bpm, resting_bpm, intensity=0.7):
    """
    Calculate the Heart Rate Reserve (HRR) training zone.

    Parameters:
    max_bpm (int or float): Maximum heart rate (beats per minute)
    resting_bpm (int or float): Resting heart rate (beats per minute)
    intensity (float): default 0.7

    Returns:
    float: Target heart rate based on HRR method
    """
    hrr_zone = (max_bpm - resting_bpm) * intensity + resting_bpm
    return round(hrr_zone, 2)



calculate_HRR(190, 60, 0.89)
```

**‚ùìQ17. Write a recursive function sum_n(n) to return sum of first n natural numbers.
Call sum_n(100).**


```python
def sum_n(n):
  if(n>0):
    return n + sum_n(n-1)   #Recursive Call
  else:
    return 0

print(sum_n(100))
```

**‚ùìQ18. Write a recursive function factorial(n) to return n!
Call factorial(6).**


```python
def factorial(n):
  if(n>0):
    return n*factorial(n-1)   #Recursive Call
  else:
    return 1;

print(factorial(5))
```

**‚ùìQ19. Write a function mean_std(array) that returns mean and standard deviation using NumPy.**

```python

def mean_std(array):
  mean = round(np.mean(array),2)  #using numpy function to find mean
  std = round(np.std(array),2)   #using numpy function to find standard deviation
  return mean,std

example = [1,5,3,2,5,6,10]
data = np.array(example)
mean,std = mean_std(data)
print("Mean is ",mean)
print("Standard deviation is ",std)
```

```python

def get_top_5_calorie_burners():
    """
    Returns the top 5 calorie burners from the dataset.

    Returns:
    pandas.DataFrame: Top 5 records with highest calories burned
    """

    # Sort by Calories_Burned in descending order and get top 5
    top_5 = df.nlargest(5, 'Calories_Burned')

    # Select relevant columns to display
    columns_to_show = ['Age', 'Gender', 'Weight (kg)', 'Calories_Burned',
                       'Workout_Type', 'Session_Duration (hours)',
                       'Name of Exercise', 'BMI']

    return top_5[columns_to_show]


# Example usage:
result = get_top_5_calorie_burners()
print("Top 5 Calorie Burners:")
print("=" * 80)
print(result.to_string(index=False))
print("\n")
```

**‚ùìQ22.Write a function to calculate and print average heart rate by gender**

```python

def calculate_avg_heart_rate_by_gender():
    """
    Calculates and prints average heart rate by gender.

    Returns:
    pandas.DataFrame: Average heart rates grouped by gender
    """

    # Calculate average heart rates by gender
    avg_heart_rate = df.groupby('Gender').agg({
        'Avg_BPM': 'mean',
    }).round(2)

    # Rename columns for better readability
    avg_heart_rate.columns = ['Avg BPM']

    # Print results
    print("=" * 70)
    print("AVERAGE HEART RATE BY GENDER")
    print("=" * 70)
    print(avg_heart_rate)
    print("\n")

    # Additional insights

    return avg_heart_rate


# Example usage:
result = calculate_avg_heart_rate_by_gender()

```

**‚ùìQ23.Write a function to visualize Calories_Burned vs Session_Duration for each Workout_Type using Matplotlib.**


```python

def plot_calories_vs_duration():
    """
    Visualizes Calories_Burned vs Session_Duration for each Workout_Type.
    """
    workout_types = df['Workout_Type'].unique()  # find all unique workout types

    plt.figure(figsize=(10, 6))

    # plot each workout type as a scatter plot
    for w_type in workout_types:
        subset = df[df['Workout_Type'] == w_type]
        plt.scatter(subset['Session_Duration (hours)'], subset['Calories_Burned'], label=w_type, s=60)

    plt.title("Calories Burned vs Session Duration by Workout Type", fontsize=14)
    plt.xlabel("Session_Duration (hours)")
    plt.ylabel("Calories Burned")
    plt.legend(title="Workout Type")
    plt.grid(True)
    plt.show()



plot_calories_vs_duration()
```

**‚ùìQ24. Write a function to compute correlation between any two numeric columns (column names taken
from user input).**


```python
def compute_correlation():
    # Show only numeric columns
    print("Numeric columns in dataset:\n", df.select_dtypes(include=[np.number]).columns.tolist())

    # Take user input for columns
    col1 = input("Enter first column name: ").strip()
    col2 = input("Enter second column name: ").strip()

    # Check if columns exist
    if col1 not in df.columns or col2 not in df.columns:
        print("‚ùå Invalid column name(s)!")
        return

    # Compute correlation
    corr_value = df[col1].corr(df[col2])

    print(f"\nCorrelation between '{col1}' and '{col2}' = {corr_value:.4f}")

compute_correlation()
```
